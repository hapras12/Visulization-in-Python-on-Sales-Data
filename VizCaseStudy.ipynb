# Question-3

# Question-6

```python
sales.head(2)
```

```python
month =sales["Month"]
```

```python
quarter = []
for x in month :
    if x in ["Jan","Feb","Mar"]:
        quarter.append("Q1")
    elif x in ["Apr","May","Jun"]:
        quarter.append("Q2")
    elif x in ["Jul","Aug","Sep"]:
        quarter.append("Q3")
    else:
        quarter.append("Q4")
```

```python
quarter
```

```python
#create a new column "Qtr"
sales["Qtr"]= pd.Series(quarter)
```

```python
sales.head(2)
```
